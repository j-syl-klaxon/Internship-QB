# Import Pandas

```python
import pandas as pd
df1 = pd.read_csv('/Users/jeffrey_sylvester/Documents/Data Folders/data_fuel_auto.csv', encoding='UTF-16 LE', delimiter='\t')
df2 = pd.read_csv('/Users/jeffrey_sylvester/Documents/Data Folders/deprecation_19_models copy.csv', encoding='Latin-1', delimiter=',')
df1.rename(columns={'Brand': 'brand'}, inplace=True) #changed the name of a columns
df1.rename(columns={'Model': 'model'}, inplace=True) #df1 and df2 now have same columns
df1.rename(columns={'Source': 'source'}, inplace=True)
df1.rename(columns={'Forecast Period': 'forecast_period'}, inplace=True)
```

```python
df2 = df2[["brand","forecast_period","model","source","modeltype","annualized_mileage","listprice","HEX(retail_amount)","trade_amount","retail_perc","trade_perc","equipment","forecast_date"]]
#arrange the columns of df2 in order
```

```python
df3=pd.merge(df1,df2, on=['brand','forecast_period','model','source'])
#combine both df1 and df2
df3.drop('Model.1',axis=1,inplace=True) #remove Model.1 column 
```

```python
df3['forecast_date'] = pd.to_datetime(df3.forecast_date)
df3.sort_values(by='forecast_date',inplace=True) #put the forecast date in order
```

```python
df3['forecast_date'] = pd.to_datetime(df3.forecast_date)
```

```python
df3["year"] = df3.forecast_date.dt.year #takes the year from forecast_date and makes it a column
```

```python
df3["month"] = df3.forecast_date.dt.month #takes the month from forecast_date and makes it a column
```
